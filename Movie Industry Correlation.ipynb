```python
# -*- coding: utf-8 -*-
"""
Created on Fri Jun 23 11:20:40 2023

@author: WX
"""
#Import libraries
from pathlib import Path

import pandas as pd 
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the configuration of the plot

# read data
DATA = Path().resolve()    
MOVIE_DATA_CSV = DATA/'movies.csv'

df = pd.read_csv(MOVIE_DATA_CSV)
```

```python
df.head()
```

```python
#checking if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{0} - {1:f}%'.format(col,pct_missing*100))
```

```python
df = df.dropna()
```

```python
#change data type for data analysis purpose
df.dtypes
```

```python
df[['budget']] = df[['budget']].astype('int64')
```

```python
df[['gross']] = df[['gross']].astype('int64')
```

```python
#extract year from the released column
df['yearcorrect'] = df['released'].str.extract(pat = '(\d{4})').astype(int) # equal to pat = '([0-9]{4})'
df['yearcorrect']
```

```python
#sort data by gross
df.sort_values(by = ['gross'],inplace = False, ascending=False)
```

```python
# drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.drop_duplicates()
```

```python
# scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
```

```python
df.head()
```

```python
#plot budget vs. gross using seaborn

sns.regplot(x='budget', y='gross', data=df, line_kws={"color":"cornflowerblue"}) 
```

```python
#correlation analysis
df.corr(method='pearson') #pearson, kendall, spearman
```

```python
#visualization of the correlation matrix

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Heatmap for the Numeric Movie Features')
```

```python
#analyse category features
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df_numerized.corr()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Heatmap for the Numeric Movie Features')
```

```python
# get the high correlation between features
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
mask = (sorted_pairs> 0.5) | (sorted_pairs < -0.5)
high_corr = sorted_pairs[mask]
high_corr
```
