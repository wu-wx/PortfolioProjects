```python
# read the data
DATA = Path().resolve()    
SC_DATA_CSV = DATA/'DataCoSupplyChainDataset.csv'

# Detect the file encoding to ensure it is read correctly
with open(SC_DATA_CSV, 'rb') as f:
    result = chardet.detect(f.read(10000)) 
    print(result)
```

```python
# Check the percentage of missing values in each column
for col in df.columns:
    val_missing = np.mean(df[col].isnull())
    print('{0} - {1:f}%'.format(col,val_missing*100))
```

```python
# Identify rows and columns that contain missing values
rows_with_na = df[df.isna().any(axis=1)]
columns_with_na = df.columns[df.isna().any()].tolist()
df_with_na = rows_with_na[columns_with_na]
print("rows and columns containing NaN values:")
print(df_with_na)
```

```python
# Drop columns with null values or unnecessary information
df=df.drop(columns=['Product Description','Product Image','Order Zipcode','Customer Password','Order Profit Per Order','Order Customer Id','Category Id','Sales per customer'])
```
